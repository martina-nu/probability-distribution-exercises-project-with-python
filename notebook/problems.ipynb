**Binomial Distribution**

In probability theory and statistics, the binomial distribution is the discrete probability distribution that gives only two possible results in an experiment, for example, when you flip a coin it can by either Head or Tails.

```python
# Probability functions, density and distribution

def dbinom(x,size,prob=0.5):
    """
    Calculates the point estimate of the binomial distribution
    """
    from scipy.stats import binom
    result=binom.pmf(k=x,n=size,p=prob,loc=0)
    return result

def pbinom(q,size,prob=0.5):
    """
    Calculates the cumulative of the binomial distribution
    """
    from scipy.stats import binom
    result=binom.cdf(k=q,n=size,p=prob,loc=0)
    return result

def qbinom(p, size, prob=0.5):
    """
    Calculates the quantile function from the binomial distribution
    """
    from scipy.stats import binom
    result=binom.ppf(q=p,n=size,p=prob,loc=0)
    return result

def rbinom(n,size,prob=0.5):
    """
    Generates random variables from the binomial distribution
    """
    from scipy.stats import binom
    result=binom.rvs(n=size,p=prob,size=n)
    return result
```

```python
dbinom(2, size = 10, prob = 0.2) #Probability that a binomial (10,0.2) takes the value 2,
pbinom(2, size = 10, prob = 0.2) #Probability that a binomial (10,0.2) will take a value less than 2
qbinom(0.9, size = 10, prob = 0.2) # what value of a binomial (10,0.2) has a cumulative probability of 0.9?
rbinom(2000, size = 10, prob = 0.2) # generate 2000 random values from a binomial distribution (10,0.2)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Discrete distributions
xs = np.arange( 1, 11 )
ys = dbinom(xs,size=10,prob=0.5) # compute the shape of the distribution
plt.plot( xs, ys, 'o' )   # plot circles...
plt.vlines( xs, 0, ys )   # ...and lines
plt.ylim( bottom=0 )      # ensure sensible bottom border
plt.title('Probability function binomial distribution')
plt.show()
```

Normal (Gaussian) Distribution

```python
#Code here
from scipy.stats import norm

#pdf: represent distributions of continuous variables with Probability Density Function

def dnorm(x,mean,sd):
    result = norm.pdf(x=x, loc=mean, scale=sd)
    return result

#cdf: The cumulative distribution function (cdf) is the probability that the variable takes a value less than or equal to x.
def pnorm(x,mean,sd):
    result = norm.cdf(x=x, loc=mean, scale=sd)
    return result

#ppf: The percent point function (ppf) is the inverse of the cumulative distribution function.  

def qnorm(probabilities,mean,sd):
    result = norm.ppf(q=probabilities, loc=mean, scale=sd)
    return result


#rvs: generate an array containing specified number of random numbers of the given normal distribution

def rnorm(mean,sd, n):
    result = norm.rvs(loc=mean, scale=sd, size = n)
    return result
```

```python
import numpy as np
import matplotlib.pyplot as plt

#Create array
x = np.arange(-10, 10, 0.1)

#pdf
y = dnorm(x,0,1) #mean= 0, standard deviation = 1

plt.title('Probability density function (pdf)')
plt.plot(x,y)
```

```python
#cdf
x = np.arange(-10, 10, 0.1)
y = pnorm(x,0,1) #mean= 0, standard deviation = 1

plt.title('Cumulative distribution function (cdf)')
plt.plot(x,y)
```

```python
#ppf

#creating an array of probability from 0 to 1
x = np.arange(0, 1, 0.01)
   
y = qnorm(x, 0, 1)
   
plt.title('Percent point function (ppf)')
plt.plot(x, y) 

#Testing different probabilities

print(qnorm(0.2, 0, 1))
print(qnorm(0.6,0, 1))
```

```python
#rvs

#creating a vector containing 1000 normally distributed random numbers

v = rnorm(0, 1, 10000)
print(v)
```

Uniform distribution


```python
from scipy.stats import uniform

#pdf: represent distributions of continuous variables with Probability Density Function

def dunif(x,a,b):
    result = uniform.pdf(x=x, loc=a, scale= b)
    return result

#cdf: The cumulative distribution function (cdf) is the probability that the variable takes a value less than or equal to x.
def punif(x,a,b):
    result = uniform.cdf(x=x, loc=a, scale=b)
    return result

#ppf: The percent point function (ppf) is the inverse of the cumulative distribution function.  

def qunif(probabilities,a,b):
    result = uniform.ppf(q=probabilities, loc=a, scale=b)
    return result


#rvs: generate an array containing specified number of random numbers of the given uniform distribution

def runif(a,b, n):
    result = uniform.rvs(loc=a, scale=b, size = n)
    return result
```

```python
#pdf
x = np.arange(-10, 10, 0.1)
   
y = dunif(x, 0, 5)
   
plt.title('Probability density function (pdf)')
plt.plot(x, y) 
```

```python
#cdf

x = np.arange(-10, 10, 0.1)
   
y = punif(x, 0, 5)
   
plt.title('Cumulative distribution function (cdf)')
plt.plot(x, y) 
```

```python
#ppf

#array of probability
x = np.arange(0, 1, 0.01)
   
y = qunif(x, 0, 5)

plt.title('Percent point function (ppf)')

plt.plot(x, y) 
```

```python
#rvs
#creating a vector containing 100 uniformly distributed random numbers

v = runif(0, 5, 100)
print(v)
```
